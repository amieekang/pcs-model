```python
import pandas as pd
```

```python
df = pd.read_csv("csv data/cleaned_modeling_features.csv")
```

```python
df.columns.tolist()
```

```python
# Identify rows that are only summaries

import re

# helper: True if this row's grade_x has ONLY S/S1/S2/S3 (any combo), False otherwise
def is_summary_only(val):
    if pd.isna(val):
        return False
    # split on comma or space
    tokens = re.split(r'[,\s]+', str(val).strip())
    tokens = [t for t in tokens if t]          # drop empty strings
    if not tokens:
        return False
    # summary-only if every token starts with 'S'
    return all(tok.startswith('S') for tok in tokens)

summary_mask = df['grade_x'].apply(is_summary_only)

# look at how many and optionally inspect them
summary_rows = df[summary_mask]
summary_rows.shape  # number of rows with only summary grades
```

```python
summary_rows[['grade_x']]
```

```python
# keep everything that is NOT summary-only
df_no_summaries = df[~summary_mask].copy()

df_no_summaries.to_csv("cleaned_modeling_features_no_summaries.csv", index=False)
```
